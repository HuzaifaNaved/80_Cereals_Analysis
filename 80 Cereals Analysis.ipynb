```python
# get min and max contents of the nutritions
nutritional_cols= ['protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','calories']
df[nutritional_cols].describe()
```

```python
# get cereals with highest and lowest nutritional values
highest_content = {}
for attribute in nutritional_cols:
    max_cereal= df.loc[df[attribute].idxmax()]['name']
    highest_content[attribute]= max_cereal
    
lowest_content = {}
for attribute in nutritional_cols:
    min_cereal= df.loc[df[attribute].idxmin()]['name']
    lowest_content[attribute]= min_cereal
```

## Working on sugars

### calculating the avg of sugars

```python
# Identify cereals with high sugar content.
# find the average sugar content, max and min sugar content
print(f'Avg Sugars:', df.sugars.mean())
print(f'Max Sugars :', df.sugars.max())
print(f'Min Sugars:', df.sugars.min())
```

```python
# lets consider range between 10 to 15 be HIGH
high_sugar= df[(df.sugars >= 10) & (df.sugars <= 15)]
high_sugar
```

```python
# categorize which manufacturer has high sugar content in them
high_sugar.mfr.value_counts()
```

```python
high_sugar.type.value_counts()
```

```python
# certainly low rating for high sugar content
high_sugar.rating.describe()
```

### high sugar content cereals which one's got the lowest rating and the highest rating

```python
# lowest one
high_sugar[high_sugar.rating == high_sugar.rating.min()]
```

```python
# highest ones
high_sugar[high_sugar.rating == high_sugar.rating.max()]
```

```python
# Find the cereals with the highest and lowest ratings
min_rating_cereal = high_sugar[high_sugar.rating == high_sugar.rating.min()]
max_rating_cereal = high_sugar[high_sugar.rating == high_sugar.rating.max()]

# Plotting with Matplotlib
fig, ax = plt.subplots(figsize=(8, 5))

# Scatter plot for all cereals
ax.scatter(high_sugar['sugars'], high_sugar['rating'], color='blue', label='All Cereals')

# Highlight the cereals with min and max ratings
ax.scatter(min_rating_cereal['sugars'], min_rating_cereal['rating'], color='red', label='Lowest Rated', s=100, edgecolor='black')
ax.scatter(max_rating_cereal['sugars'], max_rating_cereal['rating'], color='green', label='Highest Rated', s=100, edgecolor='black')

# Add labels and title
ax.set_xlabel('Sugar level')
ax.set_ylabel('Rating')
ax.set_title('High Sugar level Cereals Ratings')
ax.legend()

plt.show()
```

## Working on potass & sodium

```python
# Investigate the levels of potassium and sodium in cereals.
df[['potass','sodium']].describe()
```

## calculating the avg of potass & sodium

```python
# average potassium content is 96, so lets set the high range to be greater than 96
# average sodium content is 159 , so lets set the lower range to 0 to 159
healthy_products =df[(df.sodium <= 159) & (df.potass >= 96)]
healthy_products
```

### high sugar content cereals which one's got the lowest potass and the highest potass

```python
# Lowest ones
healthy_products[healthy_products.potass == healthy_products.potass.min()]
```

```python
# highest ones
healthy_products[healthy_products.potass == healthy_products.potass.max()]
```

```python
# Find cereals with the lowest and highest potassium content
min_potass_cereal = healthy_products[healthy_products.potass == healthy_products.potass.min()]
max_potass_cereal = healthy_products[healthy_products.potass == healthy_products.potass.max()]

# Plotting with Matplotlib
fig, ax = plt.subplots(figsize=(8, 5))

# Scatter plot for all cereals
ax.scatter(healthy_products['sugars'], healthy_products['potass'], color='blue', label='All Cereals')

# Highlight the cereals with min and max potassium
ax.scatter(min_potass_cereal['sugars'], min_potass_cereal['potass'], color='red', label='Lowest Potassium', s=100, edgecolor='black')
ax.scatter(max_potass_cereal['sugars'], max_potass_cereal['potass'], color='green', label='Highest Potassium', s=100, edgecolor='black')

# Add labels and title
ax.set_xlabel('Sugar Content')
ax.set_ylabel('Potassium Content')
ax.set_title('High Sugar Cereals Potassium Content')
ax.legend()

plt.show()
```

### high sugar content cereals which one's got the lowest sodium and the highest sodium

```python
# Lowest ones
healthy_products[healthy_products.sodium == healthy_products.sodium.min()]
```

```python
# highest ones
healthy_products[healthy_products.sodium == healthy_products.sodium.max()]
```

```python
# Find cereals with the lowest and highest sodium content
min_sodium_cereal = healthy_products[healthy_products.sodium == healthy_products.sodium.min()]
max_sodium_cereal = healthy_products[healthy_products.sodium == healthy_products.sodium.max()]

# Seaborn Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=healthy_products, x='sugars', y='sodium', color='blue', label='All Cereals')

# Highlight cereals with min and max sodium
sns.scatterplot(data=min_sodium_cereal, x='sugars', y='sodium', color='red', label='Lowest Sodium', s=150, edgecolor='black')
sns.scatterplot(data=max_sodium_cereal, x='sugars', y='sodium', color='green', label='Highest Sodium', s=150, edgecolor='black')

# Add labels and title
plt.xlabel('Sugar Content')
plt.ylabel('Sodium Content')
plt.title('High Sugar Cereals Sodium Content')
plt.legend()

plt.show()
```

### Brands & Their Ratings

```python
df_sorted = df.sort_values(by='rating', ascending=False)

plt.figure(figsize=(14, 5))
plt.title("Brands and Their Ratings")
plt.xticks(rotation = 90)
sns.barplot(data=df_sorted, x=df_sorted['name'], y=df_sorted['rating'])
```

```python
# Basic histogram with a clean layout
plt.figure(figsize=(8, 5))
sns.histplot(df['sugars'], kde=True, bins=20, color='grey', alpha=0.6)

# Add title and labels
plt.title('Sugar Content Distribution in Cereals', fontsize=14)
plt.xlabel('Sugar Content (g)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add simple threshold lines
plt.axvline(15, color='red', linestyle='--', label='High Sugar (15g+)', linewidth=1.2)
plt.axvline(5, color='green', linestyle='--', label='Low Sugar (<5g)', linewidth=1.2)

# Add legend
plt.legend()

# Display plot
plt.tight_layout()
plt.show()
```

```python
numeric_columns = df.drop(columns=['name', 'type', 'mfr','shelf', 'cups', 'weight'])

correlation_matrix = numeric_columns.corr(method = 'pearson')

# Create a heatmap for correlation with 'rating'
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix[['rating']].sort_values(by='rating', ascending=False), annot=True)
plt.title("Correlation Heatmap with Ratings")
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
